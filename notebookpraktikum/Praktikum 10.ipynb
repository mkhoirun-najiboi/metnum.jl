# Formula Kuadratur
Asumsikan bahwa $ x_k=x_0+kh $ merupakan node dengan jarak yang sama dan $ f_k=f(x_k) $. Empat formula kuadratur Newton-Cotes pertama adalah

Aturan trapesium: $ \int_{x_0}^{x_1}f(x)\,dx\approx\dfrac{h}{2}(f_0+f_1) $

Aturan Simpson: $ \int_{x_0}^{x_2}f(x)\,dx\approx\dfrac{h}{3}(f_0+4f_1+f_2) $

Aturan Simpson 3/8: $ \int_{x_0}^{x_3}f(x)\,dx\approx\dfrac{3h}{8}(f_0+3f_1+3f_2+f_3) $

Aturan Boole: $ \int_{x_0}^{x_4}f(x)\,dx\approx\dfrac{2h}{45}(7f_0+32f_1+12f_2+32f_3+7f_4) $

### Contoh 1

Diberikan fungsi 

$$  f(x) = 2 + \sin(2\sqrt{x})  $$ 

Nilai dari $ \displaystyle\int_{1}^{6}f(x)\,dx $ dapat dihampiri menggunakan 4 formula kuadratur dengan cara seperti berikut.

```julia
# Aturan Trapesium
a = 1;
b = 6;
f(x) = 2 + sin(2*sqrt(x));

h = (b-a);
# h/2(f0 + f1)
hasil1 = h/2*(f(a+0*h)+f(a+1*h))
```

```julia
# Aturan Simpson
h = (b-a)/2;
# h/3(f0+4f1+f2)
hasil2 = h/3*(f(a+0*h) + 4*f(a+1*h) + f(a+2*h))
```

```julia
# Aturan Simpson 3/8
h = (b-a)/3;
# 3h/8(f0 + 3f1 + 3f2 +f3)
hasil3 = 3*h/8*(f(a+0*h) + 3*f(a+1*h) + 3*f(a+2*h) + f(a+3*h))
```

```julia
# Aturan Boole
h = (b-a)/4;
# 2h/45(7f0 + 32f1 + 12f2 + 32f3 + 7f4)
hasil4 = 2*h/45*(7*f(a+0*h) + 32*f(a+1*h) + 12*f(a+2*h) + 32*f(a+3*h) + 7*f(a+4*h))
```

Jika diketahui nilai eksak dari
$ \displaystyle\int_{1}^{6}{f(x)\,dx}\approx 8.183479207 $, maka hampiran integral yang paling mendekati adalah aturan Boole

# Aturan Komposit
Aturan kuadratur Newton-Cotes pada sub-bab sebelumnya memiliki galat yang cukup tinggi terhadap nilai solusi eksaknya dan aturan tersebut tidak dapat mengatur galat yang dihasilkan. Untuk mengatasinya, dilakukan pembagian interval integral menjadi beberapa sub-interval. Metode ini disebut dengan aturan komposit. Pada praktikum ini, hanya akan dipelajari 2 aturan komposit, yaitu aturan komposit trapesium dan Simpson.

Aturan Komposit Trapesium: $ T(f,h)=\dfrac{h}{2}(f(a)+f(b))+h\sum_{k=1}^{M-1}f(x_k)$

Aturan Komposit Simpson: $	S(f,h)=\dfrac{h}{3}(f(a)+f(b))+\dfrac{2h}{3}\sum_{k=1}^{M-1}f(x_{2k})+\dfrac{4h}{3}\sum_{k=1}^{M}f(x_{2k-1})$

```julia
function komptrap(f,a,b,M)
  h = (b-a)/M;
  s = 0;
  # Hitung nilai sigma
  for k = 1:M-1
    x = a+k*h;
    s = s+f(x);
  end
  # Hitung hasil integral komposit trapesium
  y = h/2*(f(a)+f(b)+2*s)
end
```

```julia
function kompsimp(f,a,b,M)
  h = (b-a)/(2*M);
  s1 = 0;   s2 = 0;
  # Hitung sigma 1 yang akan dikali 4
  for k = 1:M
    x = a + (2*k-1)*h;
    s1 = s1 + f(x);
  end
  # Hitung sigma 2 yang akan dikali 2
  for k = 1:M-1
    x = a + (2*k)*h;
    s2 = s2 + f(x);
  end 
  # Hitung hasil integral komposit trapesium
  y = h/3*(f(a)+f(b)+4*s1+2*s2)
end
```

Untuk menghitung nilai dari $ \int_{1}^{6}f(x)\,dx $ dengan 11 titik contoh menggunakan aturan komposit __trapesium__ dan __Simpson__, diperlukan jumlah sub-interval sebanyak $ M=10 $. # Analisis Galat

### Contoh 4
Diberikan integral-tentu $$ \int_{1}^{6} 2 + \sin(2\sqrt{x}) \ dx $$
Berikut merupakan langkah-langkah melakukan analisis galat aturan trapesium untuk menyelesaikan integral-tentu tersebut dan menentukan kompleksitasnya menggunakan sub-interval $ M=5,\ 10,\ 20,\ 40,\ 80 $, dan $ 160 $, jika diketahui $ \int_{1}^{6}{f(x)}\,dx \approx 8.1834792077 $.

#### Aturan Komposit Trapesium

```julia
a = 1;
b = 6;
f(x) = 2+sin(2*sqrt(x));
M = [5,10,20,40,80,160];
h = (b-a)./M;
y = Array{Float64}(undef,length(M),1)
for i = 1:length(M)
    y[i] = komptrap(f,a,b,M[i]);
end
```

```julia
yeks = 8.1834792077;
err = abs.(y.-yeks);
P = [M h y err]
```

```julia
h[2]/h[1]
```

```julia
err[2:end]./err[1:end-1]
```

Karena perubahan error konvergen menuju $h^2$, kompleksitas komputasi dari aturan komposit trapesium adalah $O(h^2)$


#### Aturan Komposit Simpson

```julia
a = 1;
b = 6;
f(x) = 2+sin(2*sqrt(x));
M = [5,10,20,40,80,160];
h = (b-a)./M;
y = Array{Float64}(undef,length(M),1)
for i = 1:length(M)
    y[i] = kompsimp(f,a,b,M[i]);
end
yeks = 8.1834792077;
err = abs.(y.-yeks);
P = [M h y err]
```

```julia
h[2]/h[1]
```

```julia
err[2:end]./err[1:end-1]
```

```julia
0.5^4
```

Karena perubahan error konvergen menuju $h^4$, kompleksitas komputasi dari aturan komposit Simpson adalah $O(h^4)$
